```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('ai_adoption_dataset.csv')
data.head(20)
```

```python
data.info()
```

```python
# Check for missing values
data.isna().sum()
```

```python
# Most adopted tools overall (bar plot)
import matplotlib.pyplot as plt
import seaborn as sns   
  
```

```python
# Most adopted tools india (bar plot)
india_tools = data[data['country'] == 'India']
plt.figure(figsize=(12, 6))
order = india_tools['ai_tool'].value_counts().index
sns.countplot(data=india_tools, y='ai_tool', order=order)
plt.title('Most Adopted AI Tools Overall in India')
plt.xlabel('Count')
plt.ylabel('AI Tools')
plt.show()
```

```python
most_adopted_tools = data.groupby('ai_tool')['adoption_rate'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=most_adopted_tools.values, y=most_adopted_tools.index, palette='viridis')
plt.title("Most Adopted AI Tools (Overall)")
plt.xlabel("Total Adoption Rate")
plt.ylabel("AI Tool")
plt.tight_layout()
plt.show()
plt.savefig("Most Adopted AI Tools (Overall).png", dpi=300, bbox_inches='tight')
```

```python
abc = data.groupby('industry')['daily_active_users'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
abc.plot(kind='bar', color='blue')
plt.ylim(abc.min() - 1e6, abc.max() + 1e6)
plt.title("Total Daily Active Users by Industry")
plt.xlabel("Industry")
plt.ylabel("Total Daily Active Users")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig("daily_active_users_by_industry.png", dpi=300, bbox_inches='tight')
```

```python
country_active_users= data.groupby('country')['daily_active_users'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
country_active_users.plot(kind='bar', color='lightgreen')
plt.ylim(country_active_users.min() - 1e6, country_active_users.max() + 1e6)
plt.title("Total Daily Active Users by Country")
plt.xlabel("Country")
plt.ylabel("Total Daily Active Users")
plt.xticks(rotation=45) 
```
